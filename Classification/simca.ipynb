```python

import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
import numpy as np 
import nipal
from sklearn.model_selection import train_test_split
```

```python
#load data and separation on train and test datasets
```

```python
data = np.load('../bin_data/data.npy')
lab = np.load('../bin_data/labels.npy')
labels = np.asarray([int(i[1:])-1  for i in lab])
x_train, x_test, y_train, y_test = train_test_split(data,labels)
```

```python
#Create first model and visualization of prediction
```

```python
res1 = nipal.Nipals(pd.DataFrame(x_train[y_train == 0]))  #
res1.fit(ncomp=15)
res1.predict(pd.DataFrame(x_test))
res1.plot()
plt.show()
```

```python
#Create second model and visualization of prediction
```

```python
res2 = nipal.Nipals(pd.DataFrame(x_train[y_train != y_train[0]]), ncomp=15) #
res2.fit()
res2.predict(pd.DataFrame(x_test))
res2.plot() 
plt.show()
```
