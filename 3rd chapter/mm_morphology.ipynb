### №1 Section for importing main libraries and defining main function

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import statistics
from PIL import Image
from skimage.morphology import erosion, dilation, opening, closing, white_tophat
from skimage.morphology import disk, skeletonize

def plot2figs(image1, image2, title):
    plt.subplot(1,2,1)
    plt.axis('off')
    plt.imshow(image1, cmap=plt.cm.gray)
    plt.title('Original image')
    plt.subplot(1,2,2)
    plt.axis('off')
    plt.imshow(image2, cmap=plt.cm.gray)
    plt.title(title)
    plt.show()
```

### №2 Section for loading image

```python
image = np.asarray(Image.open(os.path.join('../bin_data', 'collagen.png')))
print('   Original image')
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

### №3 Section for morphology visualization

```python
selem = disk(3)
eroded = erosion(image, selem)
plot2figs(image, eroded, 'eroded image')
dilated = dilation(image, selem)
plot2figs(image, dilated, 'rescaled image')
opened = opening(image, selem)
plot2figs(image, opened, 'rescaled image')
skeletonized = skeletonize(opened > 128)
plot2figs(image, skeletonized, 'skeletonized image')
closed = closing(image, selem)
plot2figs(image, closed, 'closed image')
wt = white_tophat(image, selem)
plot2figs(image, wt, 'white tophat image')
```
