### №1 Section for importing main libraries and defining main function

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import statistics
from PIL import Image
from scipy.ndimage import prewitt
from skimage.transform import rescale

def plot2figs(image1, image2, title):
    plt.subplot(1,2,1)
    plt.axis('off')
    plt.imshow(image1, cmap=plt.cm.gray)
    plt.title('Original image')
    plt.subplot(1,2,2)
    plt.axis('off')
    plt.imshow(image2, cmap=plt.cm.gray)
    plt.title(title)
    plt.show()
```

### №2 Section for loading image

```python
image = Image.open(os.path.join('../bin_data', 'collagen.png'))
print('   Original image')
plt.imshow(image, cmap=plt.cm.gray)
plt.show()
```

### №3 Section for prewitt visualization

```python
rescaled = rescale(np.asarray(image), 0.25)
prewitt_img = prewitt(rescaled)
plot2figs(rescaled, prewitt_img, 'prewitt image')
```
